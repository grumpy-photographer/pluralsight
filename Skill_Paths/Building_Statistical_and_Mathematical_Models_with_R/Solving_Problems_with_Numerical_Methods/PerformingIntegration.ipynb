```r
library(mosaic)
library(Deriv)

options(repr.plot.width = 12, repr.plot.height = 8)
```

```r
f0 <- function(x) {
    m = 0; b = 3;
    m * x + b;
}

f0
```

```r
plot(f0, xlim = c(0, 5), ylim = c(0, 6), col = 'blue', lwd = 3, main = 'f(x) = m * x + b')
```

```r
I0 <- function(x) {integrate(f0, lower = 0, upper = x)$value}

vect_fun = Vectorize(I0, vectorize.args = 'x')
```

```r
t <- seq(0, 5, .5)

plot(t, vect_fun(t), xlim = c(0, 5), ylim = c(0, 20), type = 'l', lwd = 3, col = 'green', ylab = 'f(x)', xlab = 'x', main = 'f(x) = m / 2 * x^2 + b*x + c')
```

```r
f1 <- function(x) {
    2 * x + 2
}

f1
```

```r
plot(f1, xlim = range(0, 2), ylim = range(0, 10), lwd = 3, col = 'blue', ylab = 'f(x)', xlab = 'x', main = 'f(x) = 2 * x + 2')
```

```r
I1 <- function(x) {integrate(f1, lower = 0, upper = x)$value}

vect_fun = Vectorize(I1, vectorize.args='x')
```

```r
t <- seq(-5, 5, .05)

plot(t, vect_fun(t), xlim = c(-5, 5), ylim = c(-2, 15), type = 'l', lwd = 3, col = 'green', ylab = 'f(x)', xlab = 'x', main = 'f(x) = x^2 + 2*x + C')
```
