```r
library(mosaic)
library(Deriv)

options(repr.plot.width = 12, repr.plot.height = 8)
```

```r
f0 <- function(x) 4

f0
```

```r
d0 <- Deriv(f0)

d0
```

```r
f1 <- function(x) 4*x

f1
```

```r
plot(f1, xlim = c(-1, 6), ylim = c(0, 25), lwd = 4, col = 'blue')
```

```r
d1 <- Deriv(f1)

d1
```

```r
f2 <- function(x) x^2

f2
```

```r
plot(f2, xlim = c(-5, 5), ylim = c(0, 25), lwd = 4, col = 'blue')
```

```r
d2 <- Deriv(f2)

d2
```

```r
plot(d2, xlim = c(-2, 2), ylim = c(-5, 5), col = 'red', lwd = 3)
```

```r
f3 <- makeFun(4 * x ** 3 ~ x)

f3
```

```r
plotFun(f3, lwd = 4, col = 'blue')
```

```r
 d3 <- Deriv(f3)

 d3
```

```r
plot(d3, xlim = range(-2, 2), ylim = range(0, 30), lwd = 3, col = 'red')
```

```r
sin_fun <- makeFun(sin(x) ~ x)

sin_fun
```

```r
plotFun(sin_fun, xlim = range(-pi, pi), ylime = range(-1, 1), col = 'blue', lwd = 3, ylab = 'sin(x)', xlab = 'x')
```

```r
deriv_sin <- Deriv(sin_fun, 'x')

deriv_sin
```

```r
plotFun(deriv_sin, xlim = range(-pi, pi), ylim = range(-1, 1), col = 'red', lwd = 3, ylab = 'cos(x)', xlab = 'x')
```

```r
fun <- makeFun((log(x) ~ x))

fun
```

```r
plotFun(fun, xlim = range(0, 1), ylim = range(-6, 1), col = 'blue', lwd = 3, ylab = 'log(x)', xlab = 'x')
```

```r
deriv_fun <- Deriv(fun)

deriv_fun
```

```r
plotFun(deriv_fun, xlim = range(-1, 1), ylim = range(-10, 10), col = 'red', lwd = 3, ylab = '1/x', xlab = 'x') 
```
