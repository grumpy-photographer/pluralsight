```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcdefaults()
```

```python
admission_data = pd.read_csv("data/Admission_Predict.csv")
admission_data.head()
```

```python
sns.boxplot(x="GRE Score", data=admission_data)
plt.show()
```

```python
sns.boxplot(x="GRE Score", data=admission_data, orient="v")
plt.show()
```

```python
sns.boxplot(x="University Rating", y="GRE Score", data=admission_data)
plt.show()
```

```python
sns.boxplot(x="University Rating",
            y="GRE Score",
            hue="Research",
            data=admission_data,
            palette="Set1")
plt.show()
```

```python
sns.violinplot(x="GRE Score", data=admission_data)
plt.show()
```

```python
sns.violinplot(x="University Rating", y="GRE Score", data=admission_data)
plt.show()
```

```python
sns.violinplot(x="University Rating",
               y="GRE Score",
               hue="Research",
               data=admission_data,
               palette="Set2")
plt.show()
```

```python
admission_data["First Class"] = admission_data["CGPA"] > 8.5
admission_data.sample(10)
```

```python
sns.violinplot(x="University Rating", y="GRE Score", hue="First Class", data=admission_data, dodge=False)
plt.show()
```

```python
sns.boxplot(x="University Rating", y="GRE Score", data=admission_data, whis=np.inf)
sns.swarmplot(x="University Rating", y="GRE Score", data = admission_data, color="black")
plt.show()
```
