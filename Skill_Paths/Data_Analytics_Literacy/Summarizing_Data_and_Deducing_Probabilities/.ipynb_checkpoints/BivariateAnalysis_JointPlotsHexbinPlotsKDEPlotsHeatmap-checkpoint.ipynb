```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips_data = pd.read_csv("data/tips.csv")
tips_data.head()
```

```python
plt.figure(figsize=(10, 8))

sns.scatterplot(x="total_bill", y="tip", data=tips_data)

plt.xlabel("Total Bill")
plt.ylabel("Tip")

plt.show()
```

```python
plt.figure(figsize=(10, 8))

sns.scatterplot(x="total_bill", y="tip", hue="smoker", data=tips_data)

plt.xlabel("Total Bill")
plt.ylabel("Tip")

plt.show()
```

```python
plt.figure(figsize=(10, 8))

sns.scatterplot(x="total_bill",
                y="tip",
                hue="time",
                style="time",
                data=tips_data)

plt.xlabel("Total Bill")
plt.ylabel("Tip")

plt.show()
```

```python
plt.figure(figsize=(10, 8))

sns.scatterplot(x="total_bill",
                y="tip",
                hue="day",
                size="size",
                data=tips_data)

plt.xlabel("Total Bill")
plt.ylabel("Tip")

plt.show()
```

```python
plt.figure(figsize=(10, 8))

sns.scatterplot(x="total_bill",
                y="tip",
                hue="size",
                size="size",
                sizes=(30, 300),
                hue_norm=(0, 7),
                legend="full",
                data=tips_data)

plt.xlabel("Total Bill")
plt.ylabel("Tip")

plt.show()
```

```python
tips_data.plot.hexbin(x="total_bill",
                      y="tip",
                      gridsize=15,
                      figsize=(10, 8),
                      sharex=False)

plt.xlabel("Total Bill")
plt.ylabel("Tip")

plt.show()
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips_data)
plt.show()
```

```python
with sns.axes_style("white"):
    sns.jointplot(x="total_bill",
                  y="tip",
                  data=tips_data,
                  kind="hex",
                  color="r")
    plt.show()
```

```python
plt.figure(figsize=(10, 8))

sns.kdeplot(x=tips_data["total_bill"], y=tips_data["tip"])

plt.show()
```

```python
tips_corr = tips_data.corr()

tips_corr
```
