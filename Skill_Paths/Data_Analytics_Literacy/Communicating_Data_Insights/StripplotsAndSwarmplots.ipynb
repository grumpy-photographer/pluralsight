```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
mpg_data = pd.read_json("data/auto-mpg.json")
mpg_data.replace('.', np.nan, inplace=True)
mpg_data.replace('?', np.nan, inplace=True)

mpg_data = mpg_data.dropna()
mpg_data.head()
```

```python
plt.figure(figsize=(12, 8))
sns.stripplot(mpg_data["mpg"])
```

```python
plt.figure(figsize=(12, 8))
sns.stripplot(x="cylinders", y="mpg", data=mpg_data, jitter=True)
```

```python
plt.figure(figsize=(12, 8))
sns.stripplot(x="cylinders", y="mpg", data=mpg_data, jitter=True, hue="origin")
```

```python
plt.figure(figsize=(12, 8))

sns.swarmplot(x="cylinders", y="mpg", data=mpg_data)
```
