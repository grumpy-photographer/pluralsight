```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
mpg_data = pd.read_json("data/auto-mpg.json")
mpg_data.replace('.', np.nan, inplace=True)
mpg_data.replace('?', np.nan, inplace=True)

mpg_data = mpg_data.dropna()
mpg_data.head()
```

```python
 sns.lmplot(x="horsepower", y="mpg", data=mpg_data)
```

```python
sns.lmplot(x="horsepower", y="mpg", hue="origin", aspect=1.5, data=mpg_data)
```

```python
sns.lmplot(x="horsepower", y="mpg", hue="origin", palette="Set2", aspect=1.5, data=mpg_data)
```

```python
sns.lmplot(x="cylinders", y="mpg", aspect=1.5, data=mpg_data)
```

```python
sns.lmplot(x="weight", y="mpg", data=mpg_data, col="origin")
```

```python
sns.pairplot(mpg_data, height=3, vars=["mpg", "acceleration", "horsepower"],
kind="reg")
```
