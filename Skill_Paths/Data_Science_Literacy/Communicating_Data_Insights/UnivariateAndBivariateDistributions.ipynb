```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
weight_data = pd.read_json("data/weight.json")

weight_data.sample(10)
```

```python
weight_data.shape
```

```python
plt.figure(figsize=(12, 8));
sns.distplot(weight_data["Weight"], color="purple")
plt.title("Weights")
plt.ylabel("Count")
```

```python
plt.figure(figsize=(12, 8));
sns.distplot(weight_data["Weight"], kde=False, rug=True, color="purple")
plt.title("Weights")
plt.ylabel("Count")
```

```python
mpg_data = pd.read_json("data/auto-mpg.json")
mpg_data.head()
```

```python
mpg_data.replace('.', np.nan, inplace=True)
mpg_data.replace('?', np.nan, inplace=True)

mpg_data = mpg_data.dropna()
mpg_data.head()
```

```python
plt.figure(figsize=(10, 8))

sns.scatterplot(x='horsepower', y='mpg', data=mpg_data)

plt.xlabel("horsepower")
plt.ylabel("mpg")
plt.show()
```

```python
plt.figure(figsize=(10, 8))

sns.scatterplot(x="horsepower", y="mpg", data=mpg_data, hue="cylinders")

plt.xlabel("horsepower")
plt.ylabel("mpg")
plt.show()
```

```python
plt.figure(figsize=(10, 8))

sns.scatterplot(x="horsepower", y="mpg", data=mpg_data, hue="cylinders", size="weight", sizes=(30, 300), hue_norm=(0,9))

plt.xlabel("horsepower")
plt.ylabel("mpg")
plt.show()
```

```python
sns.jointplot(x="horsepower", y="mpg", data=mpg_data)
```

```python
with sns.axes_style("white"):
    sns.jointplot(x="horsepower", y="mpg", data=mpg_data, kind="hex", color="k")
```
