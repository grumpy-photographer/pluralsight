```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np_data = pd.read_csv("data/national_parks.csv")
np_data.head()
```

```python
x = np_data["Year"]
```

```python
y = np.vstack([np_data["Badlands"],
              np_data["GrandCanyon"],
              np_data["BryceCanyon"]])
y
```

```python
labels = ["Badlands", "GrandCanyon", "BryceCanyon"]
```

```python
plt.figure(figsize=(12, 8))
plt.stackplot(x, y, labels=labels)
plt.legend(loc="upper left")
plt.show()
```

```python
colors = ["sandybrown",
          "tomato",
          "skyblue"]
```

```python
plt.figure(figsize=(12, 8))

plt.stackplot(x, y,
              labels=labels,
              colors=colors,
              edgecolor="grey")

plt.legend(loc="upper left")
plt.show()
```

```python
np_data[["Badlands",
        "GrandCanyon",
         "BryceCanyon"]] = np_data[["Badlands", "GrandCanyon", "BryceCanyon"]].diff()

np_data.head()
```

```python
plt.figure(figsize=(12, 8))

plt.stem(np_data["Year"],
         np_data["GrandCanyon"],
         use_line_collection=True)
plt.title("Change in Number of Visitors")
plt.show()
```

```python
plt.figure(figsize=(12, 8))

plt.stem(np_data["Year"],
         np_data["BryceCanyon"],
         markerfmt="g_",
         linefmt="r--",
         basefmt="b:",
         use_line_collection=True)

plt.title("Change in Number of Visitors")
plt.show()
```
