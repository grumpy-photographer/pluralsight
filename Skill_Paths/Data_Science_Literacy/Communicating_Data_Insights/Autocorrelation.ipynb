```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
grand_canyon_data = pd.read_csv("data/grand_canyon_visits.csv")
grand_canyon_data.head()
```

```python
grand_canyon_data["NumVisits"].describe()
```

```python
grand_canyon_data["NumVisits"] = grand_canyon_data["NumVisits"] / 1000
grand_canyon_data["NumVisits"].describe()
```

```python
plt.figure(figsize=(12, 8))

plt.acorr(grand_canyon_data["NumVisits"], maxlags=6)
plt.show()
```

```python
plt.figure(figsize=(12, 8))

plt.acorr(grand_canyon_data["NumVisits"], maxlags=13, color="green")
plt.show()
```

```python
plt.figure(figsize=(12, 8))

plt.acorr(grand_canyon_data["NumVisits"], maxlags=48, color="green")

plt.show()
```
