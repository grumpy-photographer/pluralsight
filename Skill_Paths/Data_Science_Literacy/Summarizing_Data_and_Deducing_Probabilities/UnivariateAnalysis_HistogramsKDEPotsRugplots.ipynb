```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
height_data = pd.read_csv("data/height_data.csv")
height_data.head()
```

```python
plt.figure(figsize=(10, 8))
sns.displot(height_data["Height"], kind="kde")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.displot(height_data["Height"])
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.displot(height_data["Height"], rug=True)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.displot(height_data["Height"], rug=True, bins=40)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.rugplot(height_data["Height"], height=0.5)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.set(color_codes=True)
sns.kdeplot(height_data["Height"], shade=True)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.set(color_codes=True)
sns.kdeplot(height_data["Height"], shade=True, color="g")
plt.show()
```

```python
plt.figure(figsize=(10, 8))

sns.kdeplot(height_data["Height"], label="Height")
sns.kdeplot(height_data["Height"], bw_method=0.1, label="bw: 0.1")
sns.kdeplot(height_data["Height"], bw_method=1, label="bw: 1")

plt.legend()
plt.show()
```
