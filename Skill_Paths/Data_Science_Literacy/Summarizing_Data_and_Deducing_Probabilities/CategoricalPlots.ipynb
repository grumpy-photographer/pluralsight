```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcdefaults()
```

```python
tips_data = pd.read_csv("data/tips.csv")
tips_data.head()
```

```python
sns.barplot(x="size", y="total_bill", data=tips_data)
plt.show()
```

```python
sns.countplot(x="size", data=tips_data)
plt.show()
```

```python
sns.barplot(x="size", y="total_bill", data=tips_data, hue="smoker")
plt.show()
```

```python
sns.barplot(x="size", y="total_bill", data=tips_data, ci="sd")
plt.show()
```

```python
tips_data["couple"] = tips_data["size"] == 2
tips_data.sample(10)
```

```python
sns.barplot(x="size",
            y="total_bill",
            data=tips_data,
            hue="couple",
            dodge=False)
plt.show()
```

```python
sns.pointplot(x="size", y="total_bill", data=tips_data)
plt.show()
```

```python
sns.pointplot(x="size", y="total_bill", data=tips_data, hue="sex")
plt.show()
```

```python
sns.catplot(x="day", y="total_bill", kind="box", data=tips_data, aspect=1.5)
plt.show()
```

```python
sns.catplot(x="size",
            y="total_bill",
            hue="time",
            data=tips_data,
            height=5,
            aspect=1.5)
plt.show()
```

```python
sns.catplot(x="size",
            y="total_bill",
            hue="time",
            data=tips_data,
            kind="violin",
            height=5,
            aspect=1.5)
plt.show()
```
