```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
```

```python
brain_data = pd.read_csv("data/brain_data.csv")
brain_data.head()
```

```python
slope, intercept, r_value, _, _ = stats.linregress(
    brain_data["Head_Size"], brain_data["Brain_Weight"]
)
```

```python
print("R-square value:", r_value**2)
```

```python
slope
```

```python
intercept
```

```python
plt.figure(figsize=(8, 8))
plt.plot(
    brain_data["Head_Size"],
    brain_data["Brain_Weight"],
    "o",
    label="Original data",
    markersize=5,
)
plt.plot(
    brain_data["Head_Size"],
    slope * brain_data["Head_Size"] + intercept,
    "r",
    label="Fitted line",
)

plt.xlabel("Head Size")
plt.ylabel("Brain Weight")
plt.legend()
plt.show()
```
