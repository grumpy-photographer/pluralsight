```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
adp = pd.read_csv("datasets/automobile_data_processed.csv")

adp.sample(5)
```

```python
plt.figure(figsize=(15, 10))

fg = sns.FacetGrid(adp, col="fuel-type", height=7, aspect=1)

fg.map(plt.hist, "price", color="red")

plt.show();
```

```python
fg = sns.FacetGrid(adp, col="fuel-type", height=7, aspect=1)

fg.map(sns.scatterplot, "highway-mpg", "price", s=100)

fg.add_legend();
```

```python
fg = sns.FacetGrid(adp, col="fuel-type", hue="num-of-cylinders", height=7, aspect=1)

fg.map(sns.scatterplot, "highway-mpg", "price", s=100)

fg.add_legend();
```

```python
tips_data = pd.read_csv("datasets/tips.csv")

tips_data.sample(10)
```

```python
plt.figure(figsize=(15, 10))

fg = sns.FacetGrid(tips_data, col="size", height=7, aspect=1)

fg.map(plt.hist, "total_bill", color="green")

plt.show();
```

```python
group_size_values = np.sort(tips_data["size"].unique())
```

```python
fg = sns.FacetGrid(tips_data, row="time", height=6, aspect=1)

fg.map(sns.swarmplot, "size", "total_bill", order=group_size_values)
fg.add_legend()
```

```python
fg = sns.FacetGrid(tips_data, col="time", height=6, aspect=1)

fg.map(sns.swarmplot, "size", "total_bill", order=group_size_values)
fg.add_legend()
```
