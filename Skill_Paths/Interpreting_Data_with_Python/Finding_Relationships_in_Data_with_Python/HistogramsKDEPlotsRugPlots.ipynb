```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
auto_data = pd.read_csv("./datasets/automobile_data.csv", na_values="?")

auto_data.head()
```

```python
auto_data.shape
```

```python
auto_data.isnull().sum()
```

```python
auto_data.dropna(inplace=True)
auto_data.shape
```

```python
# auto_data.to_csv('./datasets/automobile_data_processed.csv', index=False)
```

```python
plt.figure(figsize=(12, 8))

sns.displot(auto_data["price"], color="red")

plt.title("Automobile Data")

plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.displot(auto_data["price"], color="red", bins=20)

plt.title("Automobile Data")

plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.kdeplot(auto_data["price"], color="red")

plt.title("Automobile Data")

plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.displot(auto_data["price"], color="red", rug=True)

plt.title("Automobile Data")

plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.rugplot(auto_data["price"], height=0.5, color="blue")

plt.title("Automobile Data")

plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.kdeplot(auto_data["price"], shade=True, color="blue")

plt.title("Automobile Data")

plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.scatterplot(x="horsepower", y="price", data=auto_data, s=120)

plt.title("Automobile Data")

plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.scatterplot(
    x="horsepower", y="price", data=auto_data, hue="num-of-cylinders", s=120
)

plt.title("Automobile Data")

plt.show()
```

```python
sns.regplot(x="horsepower", y="price", data=auto_data)
plt.show()
```

```python
sns.regplot(x="highway-mpg", y="price", data=auto_data)
plt.show()
```

```python
sns.jointplot(x='horsepower', y='price', data=auto_data)
plt.show()
```

```python
sns.jointplot(x="horsepower", y='price', data=auto_data, kind="reg")
plt.show()
```

```python
sns.jointplot(x='horsepower', y='price', data=auto_data, kind='kde')
plt.show()
```

```python
sns.jointplot(x="horsepower", y="price", data= auto_data, kind="hex")
plt.show()
```

```python
f, ax = plt.subplots(figsize=(6, 6))
    
sns.kdeplot(auto_data['horsepower'], auto_data['price'], ax = ax)

sns.rugplot(auto_data['horsepower'], color="limegreen", ax=ax)
sns.rugplot(auto_data["price"], color="red", vertical=True, ax=ax)

plt.title("Auto Data")

plt.show()
```
